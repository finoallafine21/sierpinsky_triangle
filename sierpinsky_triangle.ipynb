# Import all libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import random
from matplotlib.pyplot import figure
```

# All functions

```python
# this function checks if the input point (xp, yp) is within the area of the triangle (a, b, c)
def in_triangle(xp, yp, a, b, c): 

    c1 = (b[0]-a[0])*(yp-a[1])-(b[1]-a[1])*(xp-a[0])
    c2 = (c[0]-b[0])*(yp-b[1])-(c[1]-b[1])*(xp-b[0])
    c3 = (a[0]-c[0])*(yp-c[1])-(a[0]-c[1])*(xp-c[0])
    if (c1<0 and c2<0 and c3<0) or (c1>0 and c2>0 and c3>0):
        return True
        print('INSIDE')
    else:
        return False
        print('OUTSIDE')

# this function calculates the midpoint between both input points
# input points must be a list with the corresponding [x,y] coordinated
def midpoint(p1, p2):

    pn = [(p2[0]+p1[0])/2, (p2[1]+p1[1])/2]

    return pn   

#the final function, returns a list with all the points
def sierpinsky(n):
    all_points = []

    a = [1,1]
    b = [100,1]
    c = [50,100]

    while True: 

        start = [random.randint(1,100), random.randint(1,100)] #get a random point

        if in_triangle(start[0], start[1],a,b,c) == True: #make sure it is in the triangle and not outside
            starting_point = start
            break


    middle = midpoint([a,b,c][random.randint(0,2)], starting_point) #get the mid point between random corner and random start

    all_points.append(starting_point) #put the first random point into list
    all_points.append(middle) #put the first middle point into list


    #loop to calculate all the remaining points
    i=1
    while i<n:
        all_points.append(midpoint([a,b,c][random.randint(0,2)], all_points[i]))
        i=i+1

    return all_points
```

# The sierpinsky function

```python
matrix = sierpinsky(300)

j=0 #loop to plot all the points
while j<len(matrix):
    plt.plot(matrix[j][0], matrix[j][1], 'o', markersize=2)
    j=j+1
```
